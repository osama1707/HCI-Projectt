```python
import cv2
import dlib
import numpy as np
import pyaudio
import wave
import mediapipe as mp

detector = dlib.get_frontal_face_detector()
predictor = dlib.shape_predictor('C:\\Users\\Essam\\shape_predictor_68_face_landmarks.dat')
cap = cv2.VideoCapture(0)
audio = pyaudio.PyAudio()
mp_drawing = mp.solutions.drawing_utils
mp_hands = mp.solutions.hands

CHUNK = 1024
FORMAT = pyaudio.paInt16
CHANNELS = 1
SAMPLE_RATE = 44100
is_recording = False
frames = []
shape_np = []
lis = []


def calculate_distance(point1, point2):
    x1, y1 = point1
    x2, y2 = point2
    distance = np.sqrt((x2 - x1) ** 2 + (y2 - y1) ** 2)
    return distance


def audio_callback(in_data, frame_count, time_info, status):
    frames.append(in_data)
    return (in_data, pyaudio.paContinue)


flag = 0
fla=0
stream = None  # Define the 'stream' variable outside the loop
with mp_hands.Hands(
        min_detection_confidence=0.5,
        min_tracking_confidence=0.5) as hands:
    while True:
        ret, image = cap.read()
        gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
        rects = detector(gray, 1)

        if not ret:
            print("Ignoring empty camera frame.")
            continue

        image.flags.writeable = False
        results = hands.process(image)
        image.flags.writeable = True

        for rect in rects:
            shape = predictor(gray, rect)
            shape_np = np.zeros((68, 2), dtype="int")
            shape_np[62] = (shape.part(62).x, shape.part(62).y)
            shape_np[66] = (shape.part(66).x, shape.part(66).y)
            shape = shape_np

            diff = calculate_distance(shape[62], shape[66])

            if results.multi_hand_landmarks and flag == 0 and fla==0:
                for handLms in results.multi_hand_landmarks:
                    shape_np = []  
                    lis = []  
                    print('hello')
                    for id, lm in enumerate(handLms.landmark):
                        h, w, c = image.shape

                        # Finding the coordinates of each landmark
                        cx, cy = int(lm.x * w), int(lm.y * h)

                        # Printing each landmark ID and coordinates
                        # on the terminal
                        shape_np.append([cx, cy])
                        lis = shape_np
                        print(id, cx, cy)

                        # Creating a circle around each landmark
                        cv2.circle(image, (cx, cy), 10, (0,255, 0), cv2.FILLED)
                        if len(lis) >= 10:
                            diffs = calculate_distance(lis[8], lis[4])
                if diffs<110:
                    print("Start recording")
                    flag = 1
                    is_recording = False
                    stream = audio.open(format=FORMAT,
                                        channels=CHANNELS,
                                        rate=SAMPLE_RATE,
                                        input=True,
                                        frames_per_buffer=CHUNK,
                                        stream_callback=audio_callback)
                    
                    flag=1
            elif diff > 10 and flag == 0:
                print("Recording in progress")
                stream = audio.open(format=FORMAT,
                                        channels=CHANNELS,
                                        rate=SAMPLE_RATE,
                                        input=True,
                                        frames_per_buffer=CHUNK,
                                        stream_callback=audio_callback)
                flag=1
                fla=1
                is_recording = True
            if results.multi_hand_landmarks and flag==1:
                for handLms in results.multi_hand_landmarks:
                    shape_np = []  # Clear shape_np at the beginning of each hand
                    lis = []  # Clear lis at the beginning of each hand

                    for id, lm in enumerate(handLms.landmark):
                        h, w, c = image.shape

                        # Finding the coordinates of each landmark
                        cx, cy = int(lm.x * w), int(lm.y * h)

                        # Printing each landmark ID and coordinates
                        # on the terminal
                        shape_np.append([cx, cy])
                        lis = shape_np
                        print(id, cx, cy)

                        # Creating a circle around each landmark
                        cv2.circle(image, (cx, cy), 10, (0,255, 0), cv2.FILLED)
                        if len(lis) >= 10:
                            diffs = calculate_distance(lis[2], lis[9])
                            print(diffs)
                            if diffs < 90:
                                print("Stop recording")
                                is_recording = False
                                if stream is not None:
                                    stream.stop_stream()
                                    stream.close()
                                    audio.terminate()
                                    print("Recording stopped.")
                                    # Save the recorded audio frames to a WAV file
                                    wf = wave.open('C:\\Users\\Essam\\recorded_audio.wav', 'wb')
                                    wf.setnchannels(CHANNELS)
                                    wf.setsampwidth(audio.get_sample_size(FORMAT))
                                    wf.setframerate(SAMPLE_RATE)
                                    wf.writeframes(b''.join(frames))
                                    wf.close()
                                    flag = 2
                                    print("Recording saved as 'recorded_audio.wav'.")

            for i, (x, y) in enumerate(shape):
                cv2.circle(image, (x, y), 1, (0, 0, 255), -1)

        # Display the image
        cv2.imshow('Landmark Detection', image)

        # Press the escape button to terminate the code
        if cv2.waitKey(10) == 27:
            break

cap.release()
cv2.destroyAllWindows()
```
